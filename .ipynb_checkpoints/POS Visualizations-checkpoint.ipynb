```python
import pandas as pd
import glob
import os

sales_dir = '~/POS/Sales'
sales_pmt_dir = '~/POS/SalesPayments'
sales_details_dir = '~/POS/SalesDetails'

sales_csv = sorted(glob.glob(os.path.join(sales_dir, '*.csv')))
sales_pmt_csv = sorted(glob.glob(os.path.join(sales_pmt_dir, '*.csv')))
sales_details_csv = sorted(glob.glob(os.path.join(sales_details_dir, '*.csv')))

# function to read files iteratively
def read_csv(file):
    df = []
    for f in file:
        df.append(pd.read_csv(f))
    return(df)

# read in Sales file 
sales_df = read_csv(sales_csv)

# read in Sales Payments file
sales_pmt_df = read_csv(sales_pmt_csv)

# read in Sales Details file
sales_details_df = read_csv(sales_details_csv)

# read in Employee file
employees = pd.read_csv('~/POS/Employees.csv', usecols = ['First Name', 'POSID'])

# combine csv files
df_comb = []
for i in range(len(sales_df)):
    df_comb.append(sales_df[i].merge(sales_pmt_df[i], how = 'inner', on = 'Receipt Number').merge(sales_details_df[i], how = 'right', on = 'Receipt Number'))
    # parse date data
    df_comb[i]['Date_x'] = pd.to_datetime(df_comb[i]['Date_x'])
    # clean up
    df_comb[i].rename(columns = {'Amount_x': 'Receipt Total', 'Amount_y': 'Item Price', 'Date_x': 'Date'}, inplace = True)
    df_comb[i].drop(columns = ['Date_y'], inplace = True)
    df_comb[i]['Day of Week'] = df_comb[i]['Date'].dt.weekday_name
    df_comb[i]['Time of Day'] = df_comb[i]['Date'].dt.hour
    
# concatenate dataframes
data_comb = pd.concat(df_comb)
```

Plot 1: Customer Count and Average per Table by time of day and day of week

```python
count_guests = pd.DataFrame(data_comb.drop_duplicates('Receipt Number').groupby(['Day of Week', 'Time of Day'])['Number of Guests'].sum().values.reshape(3,2))
count_guests.columns = data_comb['Time of Day'].value_counts().index.values
count_guests['Day of Week'] = ['Tuesday', 'Wednesday', 'Thursday']
count_guests.set_index('Day of Week', inplace = True)

avg_guests = pd.DataFrame(data_comb.drop_duplicates('Receipt Number').groupby(['Day of Week', 'Time of Day'])['Number of Guests'].mean().values.reshape(3,2))
avg_guests.columns = data_comb['Time of Day'].value_counts().index.values
avg_guests['Day of Week'] = ['Tuesday', 'Wednesday', 'Thursday']
avg_guests.set_index('Day of Week', inplace = True)

import seaborn as sns
sns.heatmap(count_guests)

# possibly look at outliers?
```

Plot 2: Treemap of Popularity of Menu Items by Day of Week

```python
# menu item by days
menu_item_popularity = data_comb.groupby(['Day of Week', 'Time of Day', 'Menu Item'])['Quantity'].agg({'Quantity':'sum'}).groupby(by = ['Day of Week', 'Time of Day'], group_keys=False).apply(lambda x: x.sort_values(('Quantity'), ascending=False))

import squarify
import matplotlib.pyplot as plt
import matplotlib

# create square plot - for each day and each hour
menu_item_popularity = pd.DataFrame(menu_item_popularity)
```

```python
cmap = matplotlib.cm.Blues
plt.rcParams['text.color'] = 'grey'

for day in menu_item_popularity.index.levels[0].values:
    for time in menu_item_popularity.index.levels[1].values:
        plt.figure()
        subset = menu_item_popularity.loc[(day,time)]['Quantity']
        min_v=min(subset.values)
        max_v=max(subset.values)

        norm = matplotlib.colors.Normalize(vmin=min_v, vmax=max_v)
        colors = [cmap(norm(value)) for value in subset.values]
        labels = [subset.index.values[i] + '\n' + str(subset.values[i]) for i in range(len(subset.index.values))]

        squarify.plot(sizes = subset.values, label = labels, color = colors)
        plt.title('Menu Items Popularity for ' + day + ' '+ str(time))
        plt.show()
```
